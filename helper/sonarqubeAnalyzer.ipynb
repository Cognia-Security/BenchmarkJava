```python
#read results.json

import json

with open('results.json') as f:
    data = json.load(f)
```

```python
issues = data['issues']

files = set()
for issue in issues:
	files.add(issue['component'].split('/')[-1][:-5])  # remove .java
print(files)
```

```python
# read expected.csv wth header
import csv
tp_expected = []
fp_expected = []
categories = set()
with open('expected.csv') as f:
	reader = csv.DictReader(f)
	# print(len(reader.fieldnames))
	for row in reader:
		# print(row)
		if("(id:" in row['file']):
			categories.add(row['category'])
			row['id'] = row['file'].split("(id: ")[1][:-1]
			del row['file']
			if(row['real'] == 'true'):
				tp_expected.append(row)
			else:
				fp_expected.append(row)
print(tp_expected)
print(fp_expected)
print(len(fp_expected)+len(tp_expected))
print(categories)


# cwe = set()

# with open('expected.csv') as f:
# 	reader = csv.DictReader(f)
# 	print(len(reader.fieldnames))
# 	for row in reader:
# 		cwe.add(row['cwe'])
# print(list(sorted([int(i) for i in cwe])))
```

```python
filtered_tp = []
filtered_fp = []

counts = {
	"cmdi": (1,2),
	"securecookie": (0,0),
	"ldapi": (1,1),
	"pathtraver": (2,3),
	"sqli": (8,8),
	"trustbound": (0,0),
	"crypto": (2,0),
	"hash": (0,0),
	"weakrand": (0,0),
	"xpathi": (0,2),
	"xss": (6,4)
}
for category, (tp_count, fp_count) in counts.items():
	tp_added = 0
	fp_added = 0
	for tp in tp_expected:
		if tp['category'] == category:
			if tp_added < tp_count:
				filtered_tp.append(tp)
				tp_added += 1
			else:
				break
	for fp in fp_expected:
		if fp['category'] == category:
			# print(fp['category'])
			if fp_added < fp_count:
				filtered_fp.append(fp)
				fp_added += 1
			else:
				break
	print(category, tp_added, fp_added)
filtered_tp = [tp['id'] for tp in filtered_tp]
filtered_fp = [fp['id'] for fp in filtered_fp]
print("Filtered TP:")
print(filtered_tp)
print("Filtered FP:")
print(filtered_fp)
print("Total filtered issues:", len(filtered_tp) , len(filtered_fp))
together = filtered_tp+filtered_fp
print(len(set(together)))
print(together)
```
